```python
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from d2l import tensorflow as d2l
import random
```

```python
beer_df = pd.read_csv("Beer.csv")
# beer_df.info()
beer_df
```

```python
x_data = beer_df.iloc[ : , :-1].values
y_data = beer_df.iloc[ : ,  -1].values
f'dados característicos: {x_data.shape}, dado preditor: {y_data.shape}'
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Treinando o modelo
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Previsao
y_pred = model.predict(x_test)
pd.DataFrame(y_pred)
```

```python
np.set_printoptions(precision=2)
result = np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
result
```

```python
y_compare = pd.DataFrame(result)
y_compare = y_compare.rename(index=str, columns={0:'y_pred', 1:'y_test'})
y_compare = y_compare.reset_index(drop=True)
y_compare
```

```python
collumns_x = {0:'Fermentation_Time', 1:'Temperature'	, 2:'pH_Level'	, 3:'Gravity'	, 4:'Alcohol_Content'	, 5:'Bitterness'	, 6:'Color'}
x_compare = pd.DataFrame(x_test)
x_compare = x_compare.rename(index=str, columns=collumns_x)
x_compare = x_compare.reset_index(drop=True)
x_compare
```

```python
resultado_final = pd.concat([y_compare, x_compare], axis=1)
resultado_final
```
